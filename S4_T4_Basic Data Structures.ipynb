## Ints and Floats


Integers (or ints) are positive or negative whole numbers with no decimal point. Floats on the other hand represent real numbers and are written with a decimal point. The decimal point divides the number into integer and fractional parts

```python
# We do not have to explicitly specify the type. 
# the type is derived from the value being assigned.
income=50000 
print(income)

Age= 25.5
print (Age)
```

```python
# We can know the type of the object by using the type() method
print (type(income))
print (type(Age))
```

The variable 'income' is of the class integer whereas the variable 'Age' is of the class float

```python
# The result of a division operation is always of type float. 
# Hence, the variable assigned as a result of a division operation is of type float. 
Income1=50000
Income2=75000
Avg = ((Income1+Income2)/2)
print(Avg)
print(type(Avg))

```

```python
print(Avg)
```

```python
print (float(Income1)/Income2)
print (type(float(Income1)/Income2))
```

## Strings and Booleans



Strings or str is a type to represent text. bool or Boolean is a type to represent logical values. These can only take values True 
or False.

```python
Course = "Data Science with Python"
print(type(Course))
```

The class of the variable course is a string.

```python
Learning = True
print(type(Learning))
```

Since 'True' is assigned to the variable, the type of Learning is 'Boolean class',represented by bool

```python
# + (plus) operator behaves differently for strings and integers.
# For integers, plus operator results in addition of two numbers. 
3+8
```

```python
# For Strings, '+' operator concatenates strings together
'Data' + 'Science'
```

```python
# *  operator behaves differently for strings and integers.

# For Integers, the * operator performs a multiplication of two numbers.
sample2 = 1 * 3

# While working with a string, The * operator concatenates multiple copies of a string together.
sample3 = "1" * 3

print(sample2)
print(sample3)
```

```python
# '+' operator works with the objects of same class. If the operator is used on different classes 
# it results in an error. 
print("Sumit started with income of Rs." + Income1 + ". Gradually it was increased to Rs." + Income2)
```

Income1/Income2 is of type 'int' while text is of type str. Using of a '+' operator results between them, results in an error.

```python
# We can implicitly change the type of the objects so that all are of same type. 
print("Sumit started with income of Rs." + str(Income1) + ". Gradually it was increased to Rs." + str(Income2) + (". Great!"))
```

## Type conversion

Functions such as str, int, float and bool will help to convert Python values into any type

```python
# Though the value assigned is 1000, since it is within double quotes, it is treated as a string. 
Income = "10000"
type(Income)
```

```python
# To convert a string to a int - use the function int. 
NIncome = int(Income)
NIncome
```
